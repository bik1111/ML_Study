## 데이터 작업하기

### 파이토치(PyTorch)에는 데이터 작업을 위한 기본 요소 두가지인 torch.utils.data.DataLoader 와 torch.utils.data.Dataset 가 있습니다. Dataset 은 샘플과 정답(label)을 저장하고, DataLoader 는 Dataset 을 순회 가능한 객체(iterable)로 감쌉니다.

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

### torchvision.datasets 모듈은 CIFAR, COCO 등과 같은 다양한 실제 비전(vision) 데이터에 대한 Dataset을 포함한다.
### 이 튜토리얼에서는 FasionMNIST 데이터셋을 사용한다. 모든 TorchVision Dataset 은 샘플과 정답을 각각 변경하기 위한 transform 과 target_transform 의 두 인자를 포함합니다.

```python
# 공개 데이터셋에서 학습 데이터를 내려받는다
training_data = datasets.FashionMNIST(
    root = "data",
    train = True,
    download = True,
    transform = ToTensor(),
)

# 공개 데이터셋에서 테스트 데이터를 내려받는다.
test_data = datasets.FashionMNIST(
    root = "data",
    train = False,
    download = True,
    transform = ToTensor(),
)
```

### Dataset 을 DataLoader 의 인자로 전달합니다. 이는 데이터셋을 순회 가능한 객체(iterable)로 감싸고, 자동화된 배치(batch), 샘플링(sampling), 섞기(shuffle) 및 다중 프로세스로 데이터 불러오기(multiprocess data loading)를 지원한다. 

### 여기서는 배치 크기(batch size)를 64로 정의한다. 즉, 데이터로더(dataloader) 객체의 각 요소는 64개의 특징(feature)과 정답(label)을 묶음(batch)으로 반환합니다.

```python
batch_size = 64

# 데이터로더를 생성한다.
train_dataloader = DataLoader(training_data, batch_size = batch_size)
test_dataloader = DataLoader(test_data, batch_size = batch_size)

for X, y in test_dataloader:
    print(f"Shape of X [N, C, H, W]: {X.shape}")
    print(f"Shape of y: {y.shape} {y.dtype}")
    break
```

### 파일에서 사용자 정의 데이터셋 만들기


### Dataset 클래스는 반드시 3개 함수를 구현해야 합니다: __init__, __len__, and __getitem__. 아래 구현을 살펴보면 FashionMNIST 이미지들은 img_dir 디렉토리에 저장되고, 정답은 annotations_file csv 파일에 별도로 저장됩니다.



```python
import os
import pandas as pd
from torchvision.io import read_image
from torch.utils.data import Dataset

class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file, names=['file_name', 'label'])
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path)
        label = self.img_labels.iloc[idx, 1]
        
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)

        return image, label
```

#### __init__

__init__ 함수는 Dataset 객체가 생성(instantiate)될 때 한 번만 실행됩니다. 여기서는 이미지와 주석 파일(annotation_file)이 포함된 디렉토리와 (다음 장에서 자세히 살펴볼) 두가지 변형(transform)을 초기화합니다.

labels.csv 파일은 다음과 같다.

tshirt1.jpg, 0 <br>
tshirt2.jpg, 0 <br>
...... <br>
ankleboot999.jpg, 9

#### __len__

__len__ 함수는 데이터셋의 샘플 개수를 반환합니다.


#### getitem

__getitem__ 함수는 주어진 인덱스 idx에 해당하는 샘플을 데이터셋에서 불러오고 반환합니다. 인덱스를 기반으로, 디스크에서 이미지의 위치를 식별하고, read_image 를 사용하여 이미지를 텐서로 변환하고, self.img_labels의 csv 데이터로부터 해당하는 정답(label)을 가져오고, (해당하는 경우) 변형(transform) 함수들을 호출한 뒤, 텐서 이미지와 라벨을 Python 사전(dict)형으로 반환합니다.

### DataLoader로 학습용 데이터 준비하기


Dataset 은 데이터셋의 특징(feature)을 가져오고 하나의 샘플에 정답(label)을 지정하는 일을 한 번에 합니다. <br>
모델을 학습할 때, 일반적으로 샘플들을 《미니배치(minibatch)》로 전달하고, 매 에폭(epoch)마다 데이터를 다시 섞어서 과적합(overfit)을 막고, Python의 multiprocessing 을 사용하여 데이터 검색 속도를 높이려고 합니다.

DataLoader는 간단한 API로 이러한 복잡한 과정들을 추상화한 순회 가능한 객체(iterable)입니다.

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(training_data, batch_size = 64, shuffle = True)
test_dataloader = DataLoader(test_dataloader, batch_size = 64, shuffle = True)
```

아래의 각 순회(iteration)는 (각각 batch_size=64 의 특징(feature)과 정답(label)을 포함하는) train_features 와 train_labels 의 묶음(batch)을 반환합니다. <br>

## 모델 만들기

PyTorch에서 신경망 모델은 nn.Module을 상속받는 클래스(class)를 생성하여 정의합니다. __init__ 함수에서 신경망의 계층(layer)들을 정의하고 forward 함수에서 신경망에 데이터를 어떻게 전달할지 지정합니다. 가능한 경우 GPU 또는 MPS로 신경망을 이동시켜 연산을 가속(accelerate)합니다.

```python
# yTorch에서 사용 가능한 디바이스를 확인하고, 이를 출력하는 간단한 코드
device = (
    "cuda"
    # CUDA(GPU)가 사용 가능한지 여부를 확인합니다.
    if torch.cuda.is_available()
    else "mps"

    # MPS(Multi-Process Service)가 사용 가능한지 여부를 확인
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")


# 모델 정의
# nn.Module을 상속받아 신경망 모델 클래스인 NeuralNetwork를 정의합니다.
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Flatten()은 이미지를 1차원 벡터로 펼치기 위한 모듈
        self.flatten = nn.Flatten()

        # nn.Sequential을 사용하여 레이어를 순차적으로 정의
        self.linear_relu_stack = nn.Sequential(
            #입력 크기가 28*28이고 출력 크기가 512인 선형 레이어
            nn.Linear(28*28, 512),
            # sigmoid function 정의
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            # 최종 출력 레이어로, 512개의 입력을 받아 10개의 출력을 생성
            nn.Linear(512, 10)
        )
    #  순전파(forward pass) 메서드를 정의. 
    # 입력 데이터 x를 받아서 신경망을 통과시켜 로짓(logits)을 반환합니다.
    def forward(self, x):
        x = self.flatten(x)
        
        # 정의한 선형 레이어와 ReLU 활성화 함수를 순차적으로 적용
        logits = self.linear_relu_stack(x)
        return logits

model = NeuralNetwork().to(device)
print(model)


# 이 모델은 간단한 다층 퍼셉트론(MLP)으로, 
# 이미지를 펼쳐서 1차원으로 만든 후 여러 개의 선형 레이어와 활성화 함수를 거쳐 
# 최종적으로 10개의 클래스에 대한 로짓을 출력합니다.
```

## 모델 매개변수 최적화하기

모델을 학습하려면 손실 함수(loss function) 와 옵티마이저(optimizer) 가 필요합니다.

```python
# 모델의 출력(로짓)과 정답 레이블 사이의 손실을 계산합니다.
loss_fn = nn.CrossEntropyLoss()

# 확률적 경사 하강법(SGD) 최적화 알고리즘을 정의
# model.parameters()는 모델의 학습 가능한 매개변수(가중치 및 편향)를 반환하며, 이를 최적화기에 전달
# lr=1e-3는 학습률(learning rate)로, 경사 하강법 단계의 크기를 나타낸다. 
# 즉, 매개변수를 업데이트할 때 얼마나 많이 조절할지를 결정합니다.
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
```

각 학습 단계(training loop)에서 모델은 (배치(batch)로 제공되는) 학습 데이터셋에 대한 예측을 수행하고, 예측 오류를 역전파하여 모델의 매개변수를 조정합니다.


```python
def train(dataloader, mode, loss_fn, optimizer):
    size = len(dataloader.dataset)
    # 미니배치를 순회합니다. X는 입력 데이터(batch), y는 해당 입력에 대한 정답 레이블(batch)입니다.
    for batch, (X,y) in enumerate(dataloader):
        # 입력 데이터와 레이블을 GPU로 옮기는 부분입니다. to(device)를 사용하여 GPU 또는 CPU로 데이터를 전송
        X, y = X.to(device), y.to(device)

        # 예측 오류 계산
        pred = model(X)
        # 예측값과 정답 레이블 간의 손실을 계산합니다. 이 손실은 모델이 얼마나 정확한지를 나타냅니다.
        loss = loss_fn(pred,y)

        # 역전파
        # 전파 단계에서 계산된 기울기를 초기화
        # PyTorch는 각 반복에서 기울기를 누적하므로, 새로운 반복에서는 초기화해야 합니다.
        optimizer.zero_grad()
        # 손실에 대한 역전파를 수행하여 모델의 매개변수에 대한 기울기를 계산
        loss.backward()
        # 최적화기를 사용하여 모델의 매개변수를 업데이트합니다.
        optimizer.step()

        # 매 100번째 미니배치마다 현재 손실과 처리된 데이터의 양을 출력.
        if batch % 100 == 0:
            loss, current = loss.item(), (batch+1) * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


```

모델이 학습하고 있는지를 확인하기 위해 테스트 데이터셋으로 모델의 성능을 확인합니다.


```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    
    #  모델을 평가 모드로 설정
    model.eval()
    test_loss, correct = 0, 0

    # 평가 중에는 역전파를 수행하지 않음.
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            # 예측값과 정답 레이블 간의 손실을 누적
            test_loss += loss_fn(pred, y).item()
            # 정확하게 분류된 예측의 수를 누적
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    # 평가 손실에 대한 평균 계산
    test_loss /= num_batches
    # 정확도 측정.
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

학습 단계는 여러번의 반복 단계 (에폭(epochs)) 를 거쳐서 수행됩니다. 각 에폭에서는 모델은 더 나은 예측을 하기 위해 매개변수를 학습합니다. 각 에폭마다 모델의 정확도(accuracy)와 손실(loss)을 출력합니다; 에폭마다 정확도가 증가하고 손실이 감소하는 것을 보려고 합니다.

```python
epochs = 5
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)
print("Done!")
```

```python
classes = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot",
]

model.eval()
# 테스트 데이터셋에서 첫 번째 샘플의 입력과 정답을 가져옵니다.
x, y = test_data[0][0], test_data[0][1]
# torch.no_grad: 기울기 계산을 비활성화합니다. 모델을 평가할 때는 기울기가 필요하지 않으므로.
with torch.no_grad():
    x = x.to(device)
    # 모델을 사용하여 입력에 대한 예측을 수행.
    pred = model(x)
    # 예측된 클래스와 실제 클래스를 각각 변수에 할당합니다. 
    # pred[0].argmax(0)는 모델의 출력에서 가장 높은 값을 가진 클래스의 인덱스를 가져옵니다.
    predicted, actual = classes[pred[0].argmax(0)], classes[y]
    print(f'Predicted: "{predicted}", Actual: "{actual}"')
```

명시적으로 재정의(override)하지 않는다면, 인자로 주어진 텐서의 속성(모양(shape), 자료형(datatype))을 유지합니다.



```python
## PyTorch에서 생성된 텐서의 특정 속성을 유지하면서 새로운 텐서를 만드는 방법

# x_data의 속성을 유지합니다.
# x_data라는 PyTorch 텐서와 동일한 모양(shape)과 데이터 타입(dtype)을 가지면서 
# 모든 원소가 1로 채워진 새로운 텐서인 x_ones를 생성
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones} \n")

# x_data의 속성을 덮어씁니다.
# 마찬가지로 x_data와 동일한 모양을 가지면서 
# 데이터 타입이 torch.float인 랜덤한 값을 가지는 새로운 텐서 x_rand를 생성
x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Random Tensor: \n {x_rand} \n")
```

무작위(random) 또는 상수(constant) 값을 사용하기 <br>
shape 은 텐서의 차원(dimension)을 나타내는 튜플(tuple)로, 아래 함수들에서는 출력 텐서의 차원을 결정합니다.

```python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

### 텐서의 속성(Attribute)

텐서의 속성은 텐서의 모양(shape), 자료형(datatype) 및 어느 장치에 저장되는지를 나타냅니다.



```python
tensor = torch.rand(3,4)
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

NumPy식의 표준 인덱싱과 슬라이싱

```python
tensor = torch.ones(4,4)
print(tensor)
print(f"First row: {tensor[0]}")
print(f"First columm: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:, 1] = 0
print(tensor)
```

텐서 합치기 torch.cat 을 사용하여 주어진 차원에 따라 일련의 텐서를 연결할 수 있습니다. <br>



```python
# dim=1은 두 번째 차원(열 방향)을 따라서 텐서를 연결하라는 것을 의미
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)

t2 = torch.cat([tensor, tensor, tensor], dim=0)
print(t2)
```

### 산술 연산(Arithmetic operations)

```python
# 두 텐서 간의 행렬 곱(matrix multiplication)을 계산합니다. y1, y2, y3은 모두 같은 값을 갖습니다.
# ``tensor.T`` 는 텐서의 전치(transpose)를 반환합니다.

# 행렬 tensor와 그의 전치행렬을 내적한 결과를 저장합니다. @ 연산자는 행렬 곱셈을 수행
y1 = tensor @ tensor.T
# 와 동일하게 행렬 tensor와 그의 전치행렬을 내적한 결과를 저장합니다. matmul 함수는 행렬 곱셈을 수행
y2 = tensor.matmul(tensor.T)
# 행렬 y1과 동일한 크기의 무작위 값을 가진 텐서 y3를 생성
y3= torch.rand_like(y1)

# tensor와 그의 전치행렬을 내적한 결과를 y3에 저장합니다. out 매개변수는 결과를 저장할 출력 텐서를 지정
res1= torch.matmul(tensor, tensor.T, out=y3)
print(res1)

# 요소별 곱(element-wise product)을 계산합니다. z1, z2, z3는 모두 같은 값을 갖습니다.
z1 = tensor * tensor
z2 = tensor.mul(tensor)
z3 = torch.rand_like(tensor)
res2 = torch.mul(tensor, tensor, out=z3)
print(res2)
```

# 변형(Transform)

데이터가 항상 머신러닝 알고리즘 학습에 필요한 최종 처리가 된 형태로 제공되지는 않습니다. <br>
변형(transform) 을 해서 데이터를 조작하고 학습에 적합하게 만듭니다.

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root="data",
    train = True,
    download = True,
    # 이미지를 PyTorch 텐서로 변환합니다.
    transform = ToTensor(),
    # 레이블을 원-핫 인코딩으로 변환하는 함수를 정의합니다. 
    # 람다 함수는 레이블을 받아서 10개의 클래스 중 해당하는 클래스의 인덱스에 1을 할당하고, 
    # 나머지는 0으로 채운 텐서를 반환합니다.
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```
